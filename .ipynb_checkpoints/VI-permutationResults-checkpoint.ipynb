```python
#Constant Variables python
# kGroup = sys.argv[1]
# kInstrument = sys.argv[1]
# kSampleRate = 44100
# kN = 100
# kType = 'mono'
# kFeatures = 1812 #Find a way to not hard-code this.     
# path = './Music/Data/MedleyDB/Features/%s/%s/' % (sys.argv[2], kType)


#Constant Variables Jupyter
kGroup = 'tack piano'
kInstrument = 'tack piano'
kSampleRate = 44100
kN = 100 #Number of features to take into account initially.
kType = 'mono'
kFeatures = 1812 #Find a way to not hard-code this.     

#path = './Music/Data/MedleyDB/Features/no_boot_nw_2000/%s/' % (kType)

path = './Music/Data/MedleyDB/Features/Analysis/'




def dumpPickle(d, name, _path = path):
#  path = './Music/Data/MedleyDB/Features/z_Arrays/%s/' % (kType)

  with open(_path + name, 'wb') as output:
    # Pickle dictionary using protocol 0.
    pickle.dump(d, output)
    
  print '%s Saved' % (name)
  
def loadPickle(name,_path = path):  
#  path = './Music/Data/MedleyDB/Features/z_Arrays/%s/' % (kType)
  # load data from pkl file
  with open(_path + name, "rb") as fp:
      loaded_data1 = pickle.load(fp)
  
  print '%s loaded, %s ' % (name, type(loaded_data1))
  
  return loaded_data1

def plotOOBError(error, xlabel = 'n_features', ylabel = 'OOB error rate'):
    _x = []
    _y = []
    %matplotlib notebook
    
    for tupla in error:
        _x.append(tupla[0])
        _y.append(tupla[1])
    plt.rcParams['figure.figsize'] = (10,10)    
    plt.plot(_x, _y, 'ko')
    
    plt.xlim(np.min(_x), np.max(_x)+100)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(loc="upper right")
    #plt.show()
    
def plotVTscore(Vscore1, Vstd1, Vscore2, Vstd2, Tscore1, Tscore2, nFeatures, instrument):   

    %matplotlib notebook
    n_groups = 3

    fig, ax = plt.subplots()

    index = np.arange(n_groups)
    bar_width = 0.1

    opacity = 0.4
    error_config = {'ecolor': '0.3'}

    rects1 = plt.bar(index, Vscore1, bar_width,
                     alpha=opacity*2,
                     color='b',
                     yerr=Vstd1,
                     error_kw=error_config,
                     label='Validation error - %d Features' % nFeatures)

    

    rects3 = plt.bar(index + bar_width, Tscore1, bar_width,
                     alpha=opacity*2,
                     color='r',
                     label='Test error - %d Features' % nFeatures)
    
    rects2 = plt.bar(index + 2*bar_width +0.1, Vscore2, bar_width,
                     alpha=opacity,
                     color='b',
                     yerr=Vstd2,
                     error_kw=error_config,
                     label='Validation error - 3236 Features')

    rects4 = plt.bar(index + 3*bar_width +0.1, Tscore2, bar_width,
                     alpha=opacity,
                     color='r',
                     label='Test error - 3236 Features')

    plt.xlabel('Classifier')
    plt.ylabel('R^2 Score')
    plt.title('R^score Validation and Test - %s' %(instrument))
    plt.xticks(index + 4*bar_width / 2+0.05, ('RF','SV','LG'))
    plt.legend()

    plt.tight_layout()
    #plt.show()
```

```python
GInterpretationOOBError = OrderedDict()
for k in kInstruments:
    print '\n %s \n' % k.upper()
    
    _A = X[k][:]
    label = y[k]
    XPreSelected = []
    for i in gInterpretationIdx[k]:

        XPreSelected.append(_A[:,i])
    XPreSelected = np.asarray(XPreSelected).T  
    
    startTime = datetime.now() 
  
    oob_error = []
    nEstimators = 2000
    for i in range(len(gInterpretationIdx[k])):
    
        _epoch = 10
        _error = [] 
    
      
        _A = XPreSelected[:,:i+1]
        print _A.shape
    
        forest = RandomForestClassifier(n_estimators=nEstimators,
                                    warm_start=False,
                                    bootstrap = True,
                                    oob_score=True,
                                    max_features=0.3,
                                    random_state=None,
                                    n_jobs = -1)
                                                                    
        for j in range(_epoch):
        
            forest = forest.fit(_A, label)                      
            _error.append(1 - forest.oob_score_)
        
        _error_mean = np.mean(_error, axis = 0) 
        oob_error.append(_error_mean)
      
    
        print 'Batch # %d - OOB error: %f \n' % (i, _error_mean) 
  
    print '\nExecuted in: %s. \n ' % (str(datetime.now() - startTime))
    GInterpretationOOBError[k] = oob_error
```

```python
GInterpretationFeatures = OrderedDict()
GTHP = OrderedDict()
minIdx = [5,5,6,14,13]

for k,i in zip(kInstruments,minIdx):
    oob_error = GInterpretationOOBError[k]
    print '%s - Real min_oob error at idx: %d\n' % (k,np.argmin(oob_error))
    print '%s - Chosen min_oob error at idx: %d\n' % (k,i)
    
    GInterpretationFeatures[k] = gPreSelectedFeatures[k][:i+1]
    print '\n Interpretation Features: \n'
    for f in GInterpretationFeatures[k]:
        print f
    

    try:
        thresholdP = (1/(len(gPreSelectedFeatures[k])-len(GInterpretationFeatures[k]))) 
        #print thresholdP
        _th = []
        for j in np.arange(len(GInterpretationFeatures[k]),len(gPreSelectedFeatures[k])-1,1):
            _th.append(np.abs(oob_error[j+1] - oob_error[j]))
        thresholdP = thresholdP * np.sum(_th)
        #Delete this, and every k + k.upper().
        if thresholdP == 0.0:
            thresholdP = np.abs(oob_error[-1] - oob_error[-2])
    except:
        thresholdP=0.001
    print '\n THp: %f\n' % thresholdP
    GTHP[k] = thresholdP
```

```python
#Search one of the interpretation will be the first one.

GPrePredictionOOBError = OrderedDict()

for k in kInstruments:
    print '\n %s \n' % k.upper()
    
        
    _A = X[k][:]
    label = y[k]
    XPreSelected = []
    
    
    for i in gInterpretationIdx[k][0:len(GInterpretationFeatures[k])]:

        XPreSelected.append(_A[:,i])
    XPreSelected = np.asarray(XPreSelected).T  
    
    startTime = datetime.now() 
  
    oob_error = []
    nEstimators = 2000
    for i in range(len(GInterpretationFeatures[k])):
    
        _epoch = 10
        _error = [] 
    
      
        _A = XPreSelected[:,i:i+1]
        print _A.shape
    
        forest = RandomForestClassifier(n_estimators=nEstimators,
                                    warm_start=False,
                                    bootstrap = True,
                                    oob_score=True,
                                    max_features=0.3,
                                    random_state=None,
                                    n_jobs = -1)
                                                                    
        for j in range(_epoch):
        
            forest = forest.fit(_A, label)                      
            _error.append(1 - forest.oob_score_)
        
        _error_mean = np.mean(_error, axis = 0) 
        oob_error.append(_error_mean)
      
    
        print 'Batch # %d - OOB error: %f \n' % (i, _error_mean) 
  
    print '\nExecuted in: %s. \n ' % (str(datetime.now() - startTime))
    GPrePredictionOOBError[k] = oob_error
```

```python
#Re organize idx and feature vectors
GPrePredictionIdx = OrderedDict()
GPreSelectedFeatures = OrderedDict()

for k in kInstruments:
    idxP = []
    prepre = []
    
    oob_error = GPrePredictionOOBError[k]
    print '%s - min_oob error at idx: %d\n' % (k,np.argmin(oob_error))
    idxP.append(gInterpretationIdx[k][np.argmin(oob_error)])
    idx = gInterpretationIdx[k][:]
    del idx[np.argmin(oob_error)]
    for i in idx:
        idxP.append(i)
    GPrePredictionIdx[k] = idxP
    
    prepre.append(gPreSelectedFeatures[k][np.argmin(oob_error)])
    pre = gPreSelectedFeatures[k][:]
    del pre[np.argmin(oob_error)]
    for i in pre:
        prepre.append(i)
    GPreSelectedFeatures[k] = prepre
```

```python
startTime = datetime.now() 
 
GPredictionIdx = OrderedDict()
GPredictionOOBError = OrderedDict()
GPredictionOOBErrorGain = OrderedDict()

for k in kInstruments:
    print '\n %s \n' % k.upper()    
    
    idx = GPrePredictionIdx[k][:]
    thresholdP = GTHP[k]
    oob_error = []
    oob_error_total = []
    nEstimators = 2000
    predIdx = idx[:]
    errorGain = []
    _A = X[k][:]
    label = y[k]
    XPreSelected = []
    
    
    for i in gInterpretationIdx[k]:

        XPreSelected.append(_A[:,i])
    XPreSelected = np.asarray(XPreSelected).T  
    
    for i in range(len(idx)):

        _epoch = 10
        _error = [] 

        
        #_A = XPreSelected[:,:i+1]
        if i > 0:
            _A = np.hstack((_A,XPreSelected[:,i].reshape(-1,1)))
        else:
            _A = XPreSelected[:,:i+1]

        print _A.shape

        forest = RandomForestClassifier(n_estimators=nEstimators,
                                        warm_start=False,
                                        bootstrap = True,
                                        oob_score=True,
                                        max_features=0.3,
                                        random_state=None,
                                        n_jobs = -1)

        for j in range(_epoch):

            forest = forest.fit(_A, label)                      
            _error.append(1 - forest.oob_score_)

        _error_mean = np.mean(_error, axis = 0) 
        oob_error.append(_error_mean)
        oob_error_total.append(_error_mean)
        if i > 0:
            _errorGain = oob_error[-2] - oob_error[-1]

            if _errorGain > thresholdP:

                print '\n Feature - %s, Idx - %d - Added' % (GPreSelectedFeatures[k][i],i)

            else:
                _A = np.delete(_A, np.s_[-1], 1)
                predIdx.remove(idx[i])
                del oob_error[-1]

                print '\n Feature - %s, Idx - %d - Eliminated' % (GPreSelectedFeatures[k][i],i)



        else:

            _errorGain = 0
            print '\n Feature - %s, Idx - %d - Added' % (GPreSelectedFeatures[k][i],i)

        #print _errorGain            
        errorGain.append(_errorGain)   



        print '\n Batch # %d - OOB error gain: %f \n' % (i, _errorGain) 
    GPredictionIdx[k] = predIdx
    GPredictionOOBError[k] = oob_error
    GPredictionOOBErrorGain[k] = errorGain
    print '\nExecuted in: %s. \n ' % (str(datetime.now() - startTime))
```
